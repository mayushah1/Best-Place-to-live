```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
from citipy import citipy
```

```python
house_path = os.path.join('housing.csv')
house_df = pd.read_csv(house_path)
house_df.head()
```

```python
len(house_df['longitude'])
```

```python
l = len(house_df['longitude'])
row = 0
citylist = []

while row <l:
    latitude = house_df['latitude'][row]
    longitude = house_df['longitude'][row]
    avg_age = house_df['housing_median_age'][row]
    median_income = (house_df['median_income']*10000)[row]
    median_house_value = house_df['median_house_value'][row]
    
    cityname=citipy.nearest_city(latitude,longitude)
    city = cityname.city_name
    
    if city not in citylist:
        citylist.append([city,
                         latitude,
                         longitude,
                         avg_age,
                         median_income,
                         median_house_value])
    row += 1
```

```python
citylist_df = pd.DataFrame(citylist,columns={'city': city,
                                             'latitude':latitude,
                                             'longitude':longitude,
                                             'avg_age':avg_age,
                                            'median_income':median_income,
                                            'median_house_value':median_house_value})
```

```python
citylist_df.head()
```

```python
city_df = citylist_df.groupby('city')
```

```python
latitude = city_df['latitude'].mean()
longitude = city_df['longitude'].mean()
avg_age = city_df['avg_age'].mean()
avg_income = city_df['median_income'].mean()
avg_house_value = city_df['median_house_value'].mean()
```

```python
house_group = pd.DataFrame({'latitude':latitude,
                             'longitude':longitude,
                             'avg_age':avg_age,
                             'avg_income':avg_income,
                             'avg_house_value':avg_house_value})
```

```python
house_group.head()
```

```python
house_group = house_group.reset_index()
```

```python
house_group.head()
```

```python
# save the Series as a csv file
house_group.to_csv("housing.csv",index=False)
```
