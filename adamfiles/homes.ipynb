```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
from citipy import citipy
```

```python
homes_path = os.path.join('housing.csv')
homes_df = pd.read_csv(homes_path)
homes_df.head()
```

```python
len(homes_df['longitude'])
```

```python
l = len(homes_df['longitude'])
row = 0
citylist = []

while row <l:
    latitude = homes_df['latitude'][row]
    longitude = homes_df['longitude'][row]
    cityname=citipy.nearest_city(latitude,longitude)
    city = cityname.city_name
    
    if city not in citylist:
        citylist.append([city,latitude,longitude])
    row += 1
```

```python
citylist_df = pd.DataFrame(citylist,columns=['city','latitude','longitude'])
```

```python
citylist_df.head()
```

```python
merged_df = pd.merge(citylist_df,homes_df,on=['latitude','longitude'],how='outer')
merged_df
```

```python
school_path = os.path.join('school_data.csv')
school_df = pd.read_csv(school_path)
```

```python
merged_df = pd.merge(school_df,merged_df,on='city',how='outer')
merged_df.head(100)
```
